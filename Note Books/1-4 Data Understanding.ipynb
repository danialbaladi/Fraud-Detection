```python
#Import essential libraries and load the dataset
import pandas as pd

# Load the credit card fraud dataset
data_path = "/home/danial/Data Science/Fraud Detection/Data/raw/creditcard.csv"
df = pd.read_csv(data_path)

# Display the shape and the first few rows
print(f"Dataset shape: {df.shape}")
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
#Count missing values in each column
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percent
}).sort_values(by='Missing Values', ascending=False)

missing_df.head(10)
```

```python
# Quick outlier check using 'Amount' feature
import matplotlib.pyplot as plt

plt.boxplot(df['Amount'])
plt.title('Boxplot of Transaction Amount')
plt.ylabel('Amount')
plt.show()
```

```python
#Compute and inspect correlations between features
corr_matrix = df.corr()

# Show correlations with the target 'Class'
corr_with_target = corr_matrix['Class'].sort_values(ascending=False)
corr_with_target.head(10)
```

```python
#Check the class balance (Fraud vs Legit)
class_counts = df['Class'].value_counts()
class_percent = df['Class'].value_counts(normalize=True) * 100

print("Class Counts:")
print(class_counts)
print("\nClass Percentage:")
print(class_percent)
```

```python
#Optional: Visualize correlation heatmap for a subset of features
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix.iloc[:10, :10], cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap (First 10 Features)')
plt.show()
```
